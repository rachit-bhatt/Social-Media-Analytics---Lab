```python
import pandas as pd

from pytrends.request import TrendReq

import matplotlib.pyplot as plt

import numpy as np
```

```python
# Build Trend Request.
pytrend = TrendReq()
```

```python
pytrend.build_payload(kw_list=['AMD'])
```

```python
pytrend.interest_over_time()
```

```python
df = pytrend.interest_by_region()
df
```

```python
df_indexed = df.reset_index()
df_indexed
```

```python
df_sorted = df_indexed.sort_values(by='AMD', ascending = False)
df_sorted
```

```python
df_sorted.head(20).plot(kind='bar', x='geoName', y='AMD')
```
